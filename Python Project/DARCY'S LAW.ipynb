# DARCY'S LAW -- Reservoir Pressure Profile

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Darcy's Law explains the relationship between flow rate and the factors that influence it, such factors which include pressure difference,viscosity, area available to flow and permeability, The aim of this project is to study how pressure from the reservoir declines gradually until it gets to the well bore boundary

```python
## Certain assumptions have been made in order to carry out this analysis

## The reservoir is supported by an aquifer
## The reservoir is considered to be a homogeneous resevoir
## The user is meant to input certain values and where more than one values are needed it should be spaced using spcae by just clicking on the space bar
```

In this analysis radial flow would be considered since the scope of our analysis lies within the reservoir


```python
rd = float(input('\n What is the outer drainage radius of the reservoir'))
rw = float(input('\n What is the wellbore radius'))
k = float(input('\n What is the permeability of the reservoir'))
h = float(input('\n What is the thickness of the reservoir'))
nu = float(input('\n What is the viscosity of the fluid contained in the reservoir'))
B = 1
pwf = float(input('\n What is the bottom hole flowing pressure'))
q = float(input('\n What is the flow rate expected'))
```

```python
r = np.linspace(rw,rd,500)

pressure = []
for i in range(len(r)):
    p = pwf + ((141.2 * q * nu *B * (np.log(r[i]/rw)))/(k * h))
    pressure.append(p)
pressure
```

```python
plt.figure(figsize = (4,8))
```

```python
pressure = np.array(pressure)
```

```python
plt.style.use('grayscale')
```

```python
plt.figure(figsize = (4,8))
pressure = np.array(pressure)
plt.plot(r,pressure)
plt.style.use('grayscale')
plt.xlabel('radius of flow(ft)')
plt.ylabel('Pressure(psi)')
plt.title('Reservoir Pressure profile')
plt.grid()
```

From the profile, you would notice that as the radius reduces towards the wellbore the pressure drops further but steadily, this shows the well is producing a steady flow, from radius 3000 to radius 1500 you would notice a kind of balance which proves that the reservoir boundary is large it would give a fairly horizontal line, from 1500 the pressure gradually decreases which shows that the permeability of the rock is good


In summary, reservoir pressure curve is useful to petroleum engineers as it describes the exhaustion and behaviour of pressure and its impact during production.

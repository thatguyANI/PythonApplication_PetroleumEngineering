### THE KLINKENBERG EFFECT

The Klinkenberg effect explains the phenomena behind the high recorded value of gas permeability which is due to gas slippage, Klinkensberg provides a means of determining the absolute value of permeabilty using Newton Raphson method

```python
ki = input('\n What is your initial guess of the absolute permeability of the rock Ki:')
ki = float(ki)
#f(ki) = (6.9*(ki**0.64) + pm*k - pm*kg) = 0
#f'(Ki) = 4.416*(k**(-0.36)) + pm

pm = input('\n What is the measured mean pressure observed?')
pm = float(pm)

kg = input('\n What is the measured gas permeability?')
kg = float(kg)

c= 0

while (abs(6.9*(ki**0.64) + pm*ki - pm*kg) > 0.01):
    ki = ki - ((6.9*(ki**0.64) + pm*ki - pm*kg)/(4.416*(ki**(-0.36)) + pm))
    c = c + 1

print(f'the final value of absolute permeability is {ki:.3f} after {c} iterations')
```

```python
def nodal_analysis():
    import matplotlib.pyplot as plt
    print('\n Provide the data and sepate them with spaces by clicking your space bar')
    print('\n Provide Whole number values for the flow rates and pressure')
    
    q_ipr = input('\n What are the data for the flow rate into the wellbore')
    q_ipr = [int(x) for x in q_ipr.split()]
    print(q_ipr)

    pwf_ipr = input('\n What are the Pwf values for production into the wellbore?')
    pwf_ipr = [float(x) for x in pwf_ipr.split()]
    print(pwf_ipr)

    Ipr_data = {'Qipr':q_ipr,'pwf_ipr':pwf_ipr}
    x_ipr = Ipr_data['Qipr']; y_ipr = Ipr_data['pwf_ipr']

    q_tpr = input('\n What are the data for the flow rate to surface?')
    q_tpr = [int(x) for x in q_tpr.split()]
    print(q_tpr)

    p_190 = input('\n What are the PWF values for the different flowrate at 1.90 tubing diameter')
    p_190 = [int(x) for x in p_190.split()]
    print(p_190)

    p_2375 = input('\n What are the PWF values for the different flowrate at 2.375 tubing diameter')
    p_2375 = [int(x) for x in p_2375.split()]
    print(p_2375)

    p_2875 = input('\n What are the PWF values for the different flowrate at 2.875 tubing diameter')
    p_2875 = [int(x) for x in p_2875.split()]
    print(p_2875)

    tpr_data = {'Q':q_tpr,'Pwf(tbg @ 1.90)':p_190,'Pwf(tbg @ 2.375)':p_2375, 'Pwf(tbg @ 2.875)':p_2875}
    plt.figure(figsize =(8,4))
    plt.grid()
    plt.plot(x_ipr,y_ipr, label = 'IPR')
    plt.plot(tpr_data['Q'],tpr_data['Pwf(tbg @ 1.90)'], label = 'TPR for 1.90 tubing diameter', linewidth = 1.5)
    plt.plot(tpr_data['Q'],tpr_data['Pwf(tbg @ 2.375)'], label = 'TPR for 2.375 tubing diameter', linewidth = 1.5)
    plt.plot(tpr_data['Q'],tpr_data['Pwf(tbg @ 2.875)'], label = 'TPR for 2.875 tubing diameter', linewidth = 1.5)

    plt.xlabel('FlowRate(Mscf/D')
    plt.ylabel('Flowing BHP(Psia)')
    plt.title('Nodal Analysis for a Gas Well')
    plt.legend(loc ='best')
```
